**ОПИСАНИЕ ПРОЕКТА**

Сервис по продаже автомобилей с пробегом «Не бит, не крашен» разрабатывает приложение, чтобы привлечь новых клиентов. В нём можно будет узнать рыночную стоимость своего автомобиля. В вашем распоряжении данные о технических характеристиках, комплектации и ценах других автомобилей.

Критерии, которые важны заказчику:
- качество предсказания;
- время обучения модели;
- время предсказания модели. 

*Описание данных*
- DateCrawled — дата скачивания анкеты из базы
- VehicleType — тип автомобильного кузова
- RegistrationYear — год регистрации автомобиля
- Gearbox — тип коробки передач
- Power — мощность (л. с.)
- Model — модель автомобиля
- Kilometer — пробег (км)
- RegistrationMonth — месяц регистрации автомобиля
- FuelType — тип топлива
- Brand — марка автомобиля
- Repaired — была машина в ремонте или нет
- DateCreated — дата создания анкеты
- NumberOfPictures — количество фотографий автомобиля
- PostalCode — почтовый индекс владельца анкеты (пользователя)
- LastSeen — дата последней активности пользователя
- Price — цена (евро) - Целевой признак

**ЦЕЛЬ ПРОЕКТА:**
Построить модель, которая умеет определять рыночную стоимость автомобиля. Значение метрики RMSE должно быть меньше 2500.

```python
# Напишем функцию для проведения статистического анализа
def stat_data(data):
    
    # Выводим характеристики числовых признаков
    try:        
        data_describe = data.describe().T
        data_describe['var'] = data_describe['std']/data_describe['mean']*100 # Выводим коэффициент вариации
        print('Характеристики числовых признаков')
        display(data_describe)
#    except: 
#        pass # Игнорируем ошибку для датасетов без числовых признаков
    except TypeError: 
        # Выводим сообщение об ошибке, если TypeError
        print("В датасете нет числовых признаков или произошла ошибка при вычислении describe().")         
    except ZeroDivisionError:
        # Выводим сообщение об ошибке, если ZeroDivisionError
        print("Ошибка: Деление на ноль при вычислении коэффициента вариации.")          
    except Exception as e: 
        # Выводим сообщение о неожиданной ошибке
        print(f"Произошла непредвиденная ошибка: {e}")  
    
    # Проводим анализ категориальных признаков:    
    cat_col_names = data.select_dtypes(exclude=['number']).columns # Список категориальных признаков
    if len(cat_col_names) > 0:
        print('Характеристики категориальных признаков')
        print(data[cat_col_names].nunique()) # Выводим количество уникальных значений в категориальных признаках
    else:
        pass
```

```python
# Напишем функцию для построения графиков
def graph_eda(data):
    # 1. Определяем, какие столбцы являются числовыми, а какие - дискретными
    numerical_cols = data.select_dtypes(include=np.number).columns.tolist()

    # Определяем дискретные признаки (целочисленные с небольшим числом уникальных значений)
    discrete_cols = []
    for col in numerical_cols:
        if (data[col].dtype in ['int64', 'int32', 'int16', 'int8']) and (data[col].nunique() <= 35):
            discrete_cols.append(col)

    # Оставшиеся числовые признаки
    continuous_cols = [col for col in numerical_cols if col not in discrete_cols]

    # 2. Устанавливаем размер графиков по умолчанию 
    plt.rcParams["figure.figsize"] = (10, 5)

    # 3. Строим гистограммы для непрерывных числовых признаков
    if continuous_cols:
        print('Гистограммы для непрерывных числовых признаков')
        if continuous_cols:
            for i, col in enumerate(continuous_cols):
                plt.subplot(len(continuous_cols)//2 + (len(continuous_cols)%2) , 2, i + 1) # Гибкое количество строк и столбцов
                sns.histplot(data=data, x=col, kde=True)
                plt.title(f'Распределение {col}')
                plt.xticks(rotation=45) # Поворачиваем надписи на оси X для читаемости
            plt.tight_layout()
            plt.show()
    else:
        pass

    # 4. Боксплоты для числовых (непрерывных) признаков
    if continuous_cols:
        print('Боксплоты для числовых (непрерывных) признаков')
        if continuous_cols:
            for i, col in enumerate(continuous_cols):
                plt.subplot(len(continuous_cols)//2 + (len(continuous_cols)%2), 2, i + 1)
                sns.boxplot(data=data, x=col)  # Boxplot по горизонтали
                plt.title(f'Диаграмма размаха {col}')
            plt.tight_layout()
            plt.show()
    else:
        pass       
        
    # 5. Строим countplot для дискретных признаков
    if discrete_cols:
        print('Сountplot для дискретных признаков')
        if discrete_cols:
            for i, col in enumerate(discrete_cols):
                plt.subplot(len(discrete_cols)//2 + (len(discrete_cols)%2), 2, i + 1)  # Гибкое количество строк и столбцов
                sns.countplot(data=data, x=col)
                plt.title(f'Распределение {col}')
                plt.xticks(rotation=60, ha='right') # Поворачиваем надписи на оси X для читаемости
            plt.tight_layout()
            plt.show()
    else:
        pass
            
    # 6. Строим Bar plot для категориальных признаков

    # Определяем категориальные признаки
    cat_col_names = data.drop(['Model'], axis=1).select_dtypes(exclude=['number']).columns
    if len(cat_col_names) > 0:
        print('Bar plot для категориальных признаков')
        
        # Перебераем все категориальные признаки и строим графики
        for cat_col_names in cat_col_names:
#            plt.figure(figsize=(10, 10))  # Задаем размер графика для читаемости
            data[cat_col_names].value_counts().plot.bar()
            plt.title(f'{cat_col_names}')
            plt.xlabel(cat_col_names)
            plt.ylabel('Количество')
            plt.grid(True)
#            plt.legend()  
            plt.xticks(rotation=45, ha='right') # Поворот подписей оси x для лучшей читаемости
            plt.tight_layout()  # Предотвращаем обрезку подписей
            plt.show()   
    else:
        pass            
```

## Шаг 2. Обучение модели 

```python
# Напишем функцию для вычисления метрики RMSE, ее будем использовать в scoring при подборе модели
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
RANDOM_STATE = 42

# Создаём переменные для модели
X_train, X_test, y_train, y_test = train_test_split(
    df_new.drop(['Price'], axis=1),
    df_new['Price'], 
    random_state=RANDOM_STATE)

print('Размерность выборок:')
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
print()
```

### Модели линейной регрессии

```python
# Создаём списки с названиями признаков
ohe_columns = ['Brand_Model', 'VehicleType', 'FuelType', 'Repaired'] # Признаки, которые невозможно упорядочить    
ord_columns = ['Gearbox'] # Признаки, которые можно ранжировать по какому-то критерию  
num_columns = X_train.select_dtypes(include=['number']).columns

# Создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
    )

# Создаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OE
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[ 
                    ['auto', 'manual', 'unknown'],
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan # Для обработки пропусков типа ' ' 
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)

# Создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough',
    verbose_feature_names_out=False  # Упрощает имена на выходе
)

# Полный пайплайн с отбором признаков и моделью
pipe = Pipeline([
    ('preprocessor', data_preprocessor),
    ('feature_selector', SelectKBest(f_regression)),
    ('scaler', MinMaxScaler()),
    ('models', Ridge())
])
```

```python
param_grid = [
    # Словарь для модели LinearRegression()
    {
        'models': [LinearRegression()],
        'feature_selector__k': ['all', 10, 20]
    },
    
    # Словарь для модели Ridge() 
     {
        'models': [Ridge()],
        'models__alpha': np.logspace(-4, 4, 9),
#        'feature_selector__k': ['all', 10, 20],
        'scaler': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    }
    
    # Словарь для модели Lasso()
#    {
#        'models': [Lasso()],
#        'models__alpha': np.logspace(-4, 4, 9),
#        'feature_selector__k': ['all', 10, 15],
#        'scaler': [StandardScaler(), MinMaxScaler()]
#    }
]

rmse_scorer = make_scorer(rmse, greater_is_better=False) 
# greater_is_better=False потому что RandomizedSearchCV ищет МАКСИМУМ, а мы хотим МИНИМУМ RMSE

grid_search = GridSearchCV(
    pipe, 
    param_grid,
    cv=5,
    scoring=rmse_scorer,  # Используем метрику RMSE
    n_jobs=-1,
    error_score='raise'
    )

# Замер времени обучения
grid_search.fit(X_train, y_train)
print(f"Метрика лучшей модели на кросс-валидации: {-grid_search.best_score_:.2f}") # Инвертируем знак, чтобы получить положительное значение RMSE
print("Лучшая модель:", grid_search.best_estimator_)
```

```python
# Получаем результаты кросс-валидации
cv_results = grid_search.cv_results_

# Извлекаем нужные метрики для лучшей модели
best_index = grid_search.best_index_
mean_fit_time = cv_results['mean_fit_time'][best_index]
mean_score_time = cv_results['mean_score_time'][best_index]
std_test_score = cv_results['std_test_score'][best_index]

# Выводим дополнительную информацию
print("\nДополнительные метрики лучшей модели:")
print(f"Среднее время обучения: {mean_fit_time:.4f} сек")
print(f"Среднее время предсказания: {mean_score_time:.4f} сек")
print(f"Стандартное отклонение RMSE: {std_test_score:.2f}")
print("\nВсе параметры и результаты:")
for param, score in zip(cv_results['params'], -cv_results['mean_test_score']):
    print(f"Параметры: {param} | RMSE: {score:.2f}")
```

### Модель LGBMRegressor()

```python
MAX_CAT_BINS = 100  # Ограничиваем количество бинов для категориальных признаков

# Определение типов признаков
cat_columns = ['Brand_Model', 'VehicleType', 'FuelType', 'Repaired', 'Gearbox']
num_columns = X_train.select_dtypes(include=['number']).columns

# Удалить этот блок!!!!! Кодирование категориальных признаков
#encoder = OrdinalEncoder(
#    handle_unknown='use_encoded_value',
#    unknown_value=-1,
#    dtype=np.int32  # Используем int32 для экономии памяти
#)
#X_train[cat_columns] = encoder.fit_transform(X_train[cat_columns])
#X_test[cat_columns] = encoder.transform(X_test[cat_columns])

# Преобразование в категориальный тип
for col in cat_columns:
    X_train[col] = X_train[col].astype('category')
    X_test[col] = X_test[col].astype('category')

# Пайплайн для числовых признаков
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', MinMaxScaler())
])

# Общий препроцессор
preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_columns)
], remainder='passthrough')

# Определяем индексы категориальных признаков
#cat_indices = list(range(len(num_columns), len(num_columns) + len(cat_columns)))

# Полный пайплайн
lgbm_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LGBMRegressor(
        random_state=RANDOM_STATE,
        max_cat_to_onehot=5,  # Контролируем one-hot кодирование
        force_row_wise=True,  # Для ускорения обучения
        verbose=-1,  # Отключаем вывод логов
#        categorical_feature=cat_indices  # Указываем индексы категориальных признаков здесь
    ))
])

# Расширенная сетка параметров
param_grid = {
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__boosting_type': ['gbdt', 'dart'],
    'model__max_bin': [MAX_CAT_BINS]  # Ограничение для категориальных признаков
}

# Настройка GridSearch
grid_lgbm = GridSearchCV(
    estimator=lgbm_pipe,
    param_grid=param_grid,
    cv=5,
    scoring='neg_root_mean_squared_error',  # Используем встроенную метрику #make_scorer(rmse, greater_is_better=False),
    n_jobs=-1
#    verbose=2
)

# Обучение модели
grid_lgbm.fit(X_train, y_train)
print(f"Лучшие параметры: {grid_lgbm.best_params_}")
print(f"Лучший RMSE (кросс-валидация): {-grid_lgbm.best_score_:.2f}")
```

```python
# Получаем результаты кросс-валидации
cv_results_lgbm = grid_lgbm.cv_results_

# Извлекаем нужные метрики для лучшей модели
best_index = grid_lgbm.best_index_
mean_fit_time = cv_results_lgbm['mean_fit_time'][best_index]
mean_score_time = cv_results_lgbm['mean_score_time'][best_index]
std_test_score = cv_results_lgbm['std_test_score'][best_index]

# Выводим дополнительную информацию
print("\nДополнительные метрики лучшей модели:")
print(f"Среднее время обучения: {mean_fit_time:.4f} сек")
print(f"Среднее время предсказания: {mean_score_time:.4f} сек")
print(f"Стандартное отклонение RMSE: {std_test_score:.2f}")
print("\nВсе параметры и результаты:")
for param, score in zip(cv_results_lgbm['params'], -cv_results_lgbm['mean_test_score']):
    print(f"Параметры: {param} | RMSE: {score:.2f}")
```
